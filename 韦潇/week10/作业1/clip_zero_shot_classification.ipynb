# CLIP Zero-Shot 图像分类

使用ChineseCLIP模型对小狗图片进行零样本分类

```python
from PIL import Image
import matplotlib.pyplot as plt
from modelscope.pipelines import pipeline
from modelscope.utils.constant import Tasks
```

## 加载CLIP模型

```python
# 加载中文CLIP模型
pipe = pipeline(
    task=Tasks.zero_shot_image_classification,
    model='AI-ModelScope/chinese-clip-vit-base-patch16'
)
```

## 加载图片

```python
# 加载小狗图片
image_path = r"E:\AI\AI大模型课程学习\第10周：多模态大模型\Week10\小狗图片.jpg"
image = Image.open(image_path)
image
```

## 定义分类标签

```python
# 定义可能的分类标签
labels = ["一只小狗", "一只小猫", "一只兔子", "一只熊猫", "一只老虎", "一只狮子", "一只猴子", "一只大象", "一只长颈鹿", "一只鸟"]
```

## 执行Zero-Shot分类

```python
# 执行零样本分类
result = pipe(image, candidate_labels=labels)
```

## 显示分类结果

```python
# 打印分类结果
print("分类结果（按概率排序）：")
for item in result:
    print(f"{item['label']}: {item['score']:.4f}")
```

```python
# 可视化结果
plt.figure(figsize=(12, 6))

# 显示图片
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.axis('off')
plt.title('输入图片')

# 显示分类概率
plt.subplot(1, 2, 2)
labels_sorted = [item['label'] for item in result]
scores_sorted = [item['score'] for item in result]

plt.barh(range(len(labels_sorted)), scores_sorted, color='skyblue')
plt.yticks(range(len(labels_sorted)), labels_sorted)
plt.xlabel('概率')
plt.title('分类概率')
plt.tight_layout()
plt.show()
```
