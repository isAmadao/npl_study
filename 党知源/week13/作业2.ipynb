```python
from importlib.metadata import version

for p in ("huggingface_hub", "tokenizers", "safetensors", "torch"):
    print(f"{p}: {version(p)}")
```

```python
missing = []
for pkg in ("huggingface_hub", "safetensors", "torch", "tokenizers"):
    try:
        __import__(pkg)
    except ImportError:
        missing.append(pkg)
if missing:
    raise ImportError("请先安装：pip install " + " ".join(missing))
print("依赖检查通过。")
```

```python
from pathlib import Path

from huggingface_hub import hf_hub_download

REASONING_MODEL = True
REPO_ID = "Qwen/Qwen3-0.6B" if REASONING_MODEL else "Qwen/Qwen3-0.6B-Base"

local_dir = Path(REPO_ID.split("/")[-1])
local_dir.mkdir(parents=True, exist_ok=True)

print(f"仓库: {REPO_ID}")
print(f"保存到: {local_dir.resolve()}")

weights_file = hf_hub_download(
    repo_id=REPO_ID,
    filename="model.safetensors",
    local_dir=str(local_dir),
)
print("权重:", weights_file)

tok_file = hf_hub_download(
    repo_id=REPO_ID,
    filename="tokenizer.json",
    local_dir=str(local_dir),
)
print("tokenizer:", tok_file)

model_path = local_dir / "model.safetensors"
assert model_path.is_file(), model_path
size_mb = model_path.stat().st_size / (1024 * 1024)
print(f"校验通过：model.safetensors 约 {size_mb:.1f} MB")
assert (local_dir / "tokenizer.json").is_file()
```
